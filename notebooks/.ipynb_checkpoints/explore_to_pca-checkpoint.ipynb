```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

## Import local modules

```python
from modules.get_df_for_preprocessing import GetDfForPreprocessing
from modules.get_df_info import describe
from modules.prep_data_for_model import get_target
```

## Read data

```python
df = pd.read_csv('data/data.csv')
```

```python
prep = GetDfForPreprocessing(df=df)
```

```python
prep.print_df_info()
```

## Drop columns with null values above threshold

```python
non_na = prep.drop_cols_abv_na_trshld(threshold=0.3, output=True)
```

## Get statistical description of data

```python
non_na.head()
```

```python
describe(df=non_na)
```

## Convert Target column to numerical

```python
non_na['diagnosis'].unique()
```

```python
non_na['diagnosis'].replace('B',0,inplace=True)
non_na['diagnosis'].replace('M',1,inplace=True)
```

```python
non_na.head()
```

## Drop id column

```python
non_na.drop(columns=['id'], inplace=True)
```

## Get Target(y)  and Data(x)

```python
x_data, y_data = get_target(df=non_na, target_col='diagnosis')
```

```python
x_data.head()
```

```python
y_data.head()
```

## Get feature list

```python
feats = x_data.columns
```

```python
feats
```

## Normalize data(x)

```python
x_norm = pd.DataFrame(StandardScaler().fit_transform(x_data),columns=feats )
```

```python
x_norm.head()
```

## Reduce features by PCA

```python
pca = PCA(n_components=2)
```

```python
components = pca.fit_transform(x_norm)
```

```python
components_df = pd.DataFrame(components, columns=['component_1', 'component_2'])
```

```python
components_df.head()
```

## Get explantion for each component

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

## Visualize components and diagnosis

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Component_1',fontsize=20)
plt.ylabel('Component_2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = [0, 1]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = non_na['diagnosis'] == target
    plt.scatter(components_df.loc[indicesToKeep, 'component_1']
               , components_df.loc[indicesToKeep, 'component_2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
